```python
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic')
plt.rcParams['axes.unicode_minus'] =False

import pandas as pd
```

```python
def show_df(activities, values, title):

  activities = activities[:6]
  values = values[:6]

  sorted_data = sorted(zip(activities, values), key=lambda x: x[1], reverse=True)
  sorted_activities, sorted_values = zip(*sorted_data)

  plt.figure(figsize=(10, 6))
  plt.bar(sorted_activities, sorted_values)
  plt.title(title)
  plt.xlabel("활동")
  plt.ylabel("선호도 (%)")
  plt.xticks(rotation=45)
  plt.tight_layout()
  plt.show()
```

# 1. 문화예술 관람 활동

```python
activities = ["영화관람", "미술관 관람", "박물관 관람", "연예공연 관람", "연극공연 관람", "음악연주회 관람", "무용공연 관람", "전통예술공연 관람"]
values = [74, 12.5, 11.5, 8.4, 7.1, 4, 2.6, 1.8]
title = "1. 문화예술 관람 활동"

show_df(activities, values, title)
```
